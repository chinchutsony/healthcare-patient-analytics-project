```python
import pandas as pd
```

```python
# Load dataset
df = pd.read_csv('/content/healthcare_patient_analytics_seaborn 2.csv')

# Show first 5 rows
df.head()
```

#Department Analysis


```python
import matplotlib.pyplot as plt

dept_counts = df['department'].value_counts()

plt.figure()
dept_counts.plot(kind='bar')
plt.title("Patients by Department")
plt.xlabel("Department")
plt.ylabel("Number of Patients")
plt.show()
```

# Treatment Cost Trends

```python
df['treatment_cost'].plot(kind='hist', bins=20)
plt.title("Treatment Cost Distribution")
plt.xlabel("Cost")
plt.show()
```

```python
df['recovery_score'].plot(kind='hist', bins=20)
plt.title("Recovery Score Distribution")
plt.xlabel("Recovery Score")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Create categories
df['Risk_Category'] = pd.cut(
    df['readmission_risk'],
    bins=[0, 0.3, 0.6, 1.0],
    labels=['Low', 'Medium', 'High']
)

# Count categories
risk_counts = df['Risk_Category'].value_counts()

# Create pie chart
plt.figure()
risk_counts.plot(kind='pie', autopct='%1.1f%%')

plt.title("Readmission Risk Levels")
plt.ylabel("")

plt.show()
```
